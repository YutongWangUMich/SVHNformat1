```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Path to your folder containing images
image_folder = './data_proc/test/'

def plot_images(image_folder, grid_size=(8, 8)):
    fig, axs = plt.subplots(grid_size[0], grid_size[1], figsize=(8, 8))

    # Get all png files from the image folder
    image_files = sorted(
        (f for f in os.listdir(image_folder) if f.endswith('.png')),
        key=lambda x: int(x.split('.')[0]))  # sort files numerically based on the filename before the extension

    # Loop through the images and plot them to the grid
    for i, ax in enumerate(axs.flat):
        if i >= len(image_files) or i >= grid_size[0] * grid_size[1]:
            break  # stop if there are no more images or grid is full
        img = mpimg.imread(os.path.join(image_folder, image_files[i]))
        ax.imshow(img)
        ax.axis('off')  # turn off axes for clarity

    plt.show()

plot_images(image_folder)
```
